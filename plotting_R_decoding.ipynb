```r
library(tidyverse)
library(cowplot)
library(ragg)
library(RcppCNPy)
```

```r
x = c(sqrt(3/4), 0, -sqrt(3/4), -sqrt(3/4), 0, sqrt(3/4),sqrt(3/4))
y = c(1/2, 1, 1/2, -1/2,-1,-1/2,1/2)

xyoo = npyLoad('/Users/weiwei/BaiduCloud/Work/Paris/HSV/Results_167_nnls/figs/vector/f_06/mesh.npy')
xyoo.color=npyLoad('/Users/weiwei/BaiduCloud/Work/Paris/HSV/Results_167_nnls/figs/vector/f_06/mesh_color.npy')

x_all = c(x, x*0.8, x*0.6, x*0.4, x*0.2, c(0,x[2],0,x[3],0,x[4],0,x[5],0,x[6]))
y_all = c(y, y*0.8, y*0.6, y*0.4, y*0.2, c(0,y[2],0,y[3],0,y[4],0,y[5],0,y[6]))

df.hex = data.frame(x_all,y_all)
colnames(df.hex) = c('x','y')

df.mesh = data.frame(xyoo, rgb(xyoo.color))
colnames(df.mesh) = c('xoo','yoo','colors')
```

```r
xyoo.color
```

```r
options(repr.plot.width = 5, repr.plot.height = 5.8)

hex.big = ggplot(data=df.mesh, aes(xoo,yoo)) +
            geom_path(data=df.hex, alpha=0.5, linewidth=0.2, aes(x,y)) +
            geom_point(color=df.mesh$colors, size=3, alpha=0.6) +
            theme_void() +
            theme(legend.position='none')

print(hex.big)
```
